```python
import matplotlib.pyplot as plt
import seaborn as sns

fig, axs = plt.subplots(2, 2, figsize=(12, 10))

# Complains
data.groupby('Complains').size().plot(kind='barh', color=sns.palettes.mpl_palette('plasma'), ax=axs[0, 0])
axs[0, 0].set_title('Complains')
axs[0, 0].spines[['top', 'right']].set_visible(False)

# Plan
data.groupby('Plan').size().plot(kind='barh', color=sns.palettes.mpl_palette('copper'), ax=axs[0, 1])
axs[0, 1].set_title('Plan')
axs[0, 1].spines[['top', 'right']].set_visible(False)

# Status
data.groupby('Status').size().plot(kind='barh', color=sns.palettes.mpl_palette('Wistia'), ax=axs[1, 0])
axs[1, 0].set_title('Status')
axs[1, 0].spines[['top', 'right']].set_visible(False)

# Churn
data.groupby('Churn').size().plot(kind='barh', color=sns.palettes.mpl_palette('Dark2'), ax=axs[1, 1])
axs[1, 1].set_title('Churn')
axs[1, 1].spines[['top', 'right']].set_visible(False)
plt.tight_layout()

plt.show()
```

```python
import matplotlib.pyplot as plt

fig, axes = plt.subplots(1, 2, figsize=(12, 6))

# Boxplot untuk Charge Amount
axes[0].boxplot(data['Charge Amount'])
axes[0].set_title('Charge Amount')

# Boxplot untuk Age
axes[1].boxplot(data['Age'])
axes[1].set_title('Age')

plt.tight_layout()
plt.show()
```

```python
plt.scatter(data['Freq. of use'], data['Freq. of SMS'])
plt.title('Freq. of use vs Freq. of SMS')
plt.xlabel('Freq. of use')
plt.ylabel('Freq. of SMS')
plt.show()
```
