#Rangkuman Matplotlib

`Matplotlib` adalah pustaka plotting yang hebat dalam Python yang digunakan untuk membuat visualisasi statistik, animasi, dan interaktif. Tujuan utama Matplotlib adalah menyediakan alat dan fungsi bagi pengguna untuk merepresentasikan data secara grafis, sehingga lebih mudah untuk dianalisa dan dipahami. Awalnya dikembangkan oleh John D. Hunter pada tahun 2003 dan sekarang dikelola oleh komunitas pengembang yang besar.

`Keunggulan Utama Matplotlib`:
1. `Fleksibilitas` : Matplotlib dapat menghasilkan berbagai macam plot, termasuk plot garis, plot sebar, plot batang, histogram, diagram lingkaran, dan masih banyak lagi.
2. `Kustomisasi` : menyediakan opsi kustomisasi yang luas untuk mengontrol setiap aspek plot, seperti gaya garis, warna, penanda, label, dan anotasi.
3. `Integrasi dengan NumPy` : Matplotlib terintegrasi mulus dengan NumPy, memudahkan untuk memplot data array secara langsung.
4. `Kualitas Publikasi` : Matplotlib menghasilkan plot berkualitas tinggi yang cocok untuk publikasi dengan kontrol ketat atas estetika plot.
5. `Dapat mencakup` : Matplotlib sangat dapat diinstal, dengan ekosistem besar perangkat tambahan dan ekstensi seperti Seaborn, fungsi pembuatan grafik Pandas, dan Basemap untuk pembuatan grafik geografis.
6. `Lintas-Platform` : Bersifat independen terhadap platform dan dapat berjalan pada berbagai sistem operasi, termasuk Windows, macOS, dan Linux.
7. `Plot Interaktif` : Matplotlib mendukung plot interaktif melalui penggunaan widget dan penanganan peristiwa, yang memungkinkan pengguna menjelajahi data secara dinamis.

```python
# 1. Instalasi & Import
#pip install matplotlib
import matplotlib.pyplot as plt
```

```python
# 2. Struktur Dasar Matplotlib
# 2.1 Pyplot Interface (plt.plot(), plt.xlabel(), plt.ylabel(), plt.title(), dan sebagainya.)
x = [1, 2, 3, 4]
y = [1, 4, 9, 16]

plt.plot(x, y)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Simple Line Plot')
plt.show()
```

```python
# 2.2 Object-Oriented Interface (membuat objek figure dan axis untuk memberikan kontrol lebih detail.)
fig, ax = plt.subplots()
ax.plot(x, y)
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_title('Simple Line Plot')
plt.show()
```

```python
# 3. Jenis-Jenis Plot dalam Matplotlib
# 3.1 Line Plot
plt.plot([1, 2, 3, 4], [1, 4, 9, 16])
plt.show()
```

```python
# 3.2 Scatter Plot
x = [1, 2, 3, 4, 5]
y = [5, 7, 4, 3, 9]
plt.scatter(x, y)
plt.show()
```

```python
# 3.3 Bar Plot
categories = ['A', 'B', 'C', 'D']
values = [3, 7, 5, 9]
plt.bar(categories, values)
plt.show()
```

```python
# 3.4 Histogram
data = [1, 2, 2, 3, 3, 3, 4, 4, 5]
plt.hist(data, bins=5)
plt.show()
```

```python
# 3.5 Pie Chart
labels = ['A', 'B', 'C', 'D']
sizes = [15, 30, 45, 10]
plt.pie(sizes, labels=labels)
plt.show()
```

```python
# 4. Pengaturan Tampilan Plot
# 4.1 Label Axis
plt.xlabel('X Axis')
plt.ylabel('Y Axis')

# 4.2 Label Tittle
plt.title('Title of the Plot')

# 4.3 Label Legend
plt.plot(x, y, label='Data 1')
plt.legend()

# 4.4 Style
plt.style.use('ggplot')  # Menggunakan style 'ggplot'
```

```python
# 5. Subplot
fig, (ax1, ax2) = plt.subplots(1, 2)
ax1.plot(x, y)
ax2.bar(categories, values)
plt.show()
```

```python
# 6. Visualisasi 3D
from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter([1, 2, 3], [4, 5, 6], [7, 8, 9])
plt.show()
```

```python
# 7. Menyimpan Gambar
plt.plot(x, y)
plt.savefig('plot.png')
```

```python
# 8. Integrasi dengan Pandas
import pandas as pd
df = pd.DataFrame({
    'A': [1, 2, 3, 4],
    'B': [5, 6, 7, 8]
})
df.plot(kind='line')
plt.show()
```

```python
# 9. Animasi Dengan Matpplotlib
from matplotlib.animation import FuncAnimation

fig, ax = plt.subplots()
xdata, ydata = [], []
ln, = plt.plot([], [], 'r')

def init():
    ax.set_xlim(0, 10)
    ax.set_ylim(0, 10)
    return ln,

def update(frame):
    xdata.append(frame)
    ydata.append(frame ** 2)
    ln.set_data(xdata, ydata)
    return ln,

ani = FuncAnimation(fig, update, frames=range(10), init_func=init)
plt.show()
```

#Rangkuman Seaborn

`Seaborn` adalah pustaka visualisasi yang menakjubkan untuk membuat grafik statistik dalam Python. Pustaka ini menyediakan gaya dan palet warna default yang cantik untuk membuat grafik statistik lebih menarik. Pustaka ini dibangun di atas pustaka matplotlib dan juga terintegrasi erat dengan struktur data dari pandas.

Seaborn bertujuan untuk menjadikan visualisasi sebagai bagian utama dari penjelajahan dan pemahaman data. Pustaka ini menyediakan API berorientasi kumpulan data sehingga kita dapat beralih di antara representasi visual yang berbeda untuk variabel yang sama guna mendapatkan pemahaman yang lebih baik tentang kumpulan data.

`Keunggulan Utama Seaborn`:
1. `Visualisasi statistik`: Seaborn merupakan alat yang ideal untuk memvisualisasikan hubungan dalam data.
2. `Sintaksis sederhana`: Seaborn memiliki sintaksis yang intuitif sehingga memungkinkan pembuatan visualisasi kompleks dengan kode minimal.
3. `Tema dan estetika bawaan`: Seaborn memiliki beberapa tema dan palet warna bawaan yang dapat meningkatkan daya tarik visual plot.
4. `Integrasi dengan pandas`: Seaborn terintegrasi dengan baik dengan kerangka data pandas.
5. `Plot interaktif dan animasi`: Seaborn mendukung plot interaktif dan animasi dengan bantuan pustaka lain seperti plotly, bokeh, dan ipywidgets.
6. `Analisis statistik tingkat lanjut`: Seaborn menawarkan berbagai fungsi statistik bawaan yang memungkinkan pengguna untuk melakukan analisis statistik yang rumit.
7. `Visualisasi multiplot`: Seaborn memungkinkan pengguna untuk membuat visualisasi multiplot yang kompleks, seperti kisi-kisi plot.
8. `Komunitas pengguna yang aktif`: Komunitas pengguna Seaborn yang aktif menyediakan akses ke teknik dan tren terbaru.

```python
# 1. Instalasi dan Import
#pip install seaborn
import seaborn as sns
```

```python
# 2. Grafik Distribusi
data = [1, 2, 2, 3, 3, 3, 4, 5, 6, 6, 7, 8, 9]
sns.histplot(data, kde=True)
plt.title('Histogram dengan KDE')
plt.show()
```

```python
# 3. Scatter Plot
tips = sns.load_dataset("tips")
sns.scatterplot(x='total_bill', y='tip', data=tips)
plt.title('Scatter Plot antara Total Bill dan Tip')
plt.show()
```

```python
# 4. Relasi Plot
sns.relplot(x="total_bill", y="tip", hue="sex", data=tips)
plt.title('Relational Plot dengan Hue')
plt.show()
```

```python
# 5. Bar dan Count Plot
sns.barplot(x='day', y='total_bill', data=tips)
plt.title('Bar Plot Total Bill per Hari')
plt.show()

sns.countplot(x='day', data=tips)
plt.title('Count Plot Jumlah Pesanan per Hari')
plt.show()
```

```python
# 6. Boxplot
sns.boxplot(x='day', y='total_bill', data=tips)
plt.title('Box Plot Total Bill per Hari')
plt.show()
```

```python
# 7. Heatmap
numerical_tips = tips.select_dtypes(include=['number'])
corr = numerical_tips.corr()
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('Heatmap Korelasi')
plt.show()
```

```python
# 8. Pair Plot
sns.pairplot(tips)
plt.show()
```

```python
# 9. Menambahkan warna dan styling
sns.set_style('whitegrid')
sns.scatterplot(x='total_bill', y='tip', data=tips, hue='sex', palette='coolwarm')
plt.title('Scatter Plot dengan Gaya dan Warna')
plt.show()
```

```python
# 10. Violin Plot
sns.violinplot(x='day', y='total_bill', data=tips)
plt.title('Violin Plot Total Bill per Hari')
plt.show()
```

```python
# 11. Subplot dengan FacetGrid
g = sns.FacetGrid(tips, col='sex')
g.map(sns.scatterplot, 'total_bill', 'tip')
plt.show()
```

```python
# 12. Menyimpan Grafik
sns.barplot(x='day', y='total_bill', data=tips)
plt.savefig('barplot.png')
```
